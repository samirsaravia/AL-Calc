Quando o gás hidrogênio reage com iodo gasoso a temperaturas elevadas, é estabelecido o seguinte equilibrio:
$$H_{2(g)}+I_{2(g)} \leftrightarrows HI_{(g)}$$

Um estudante mediu a constante de equilibrio como 59,3 a 400˚C , se alguem tentar começar com uma mistura que inclui $0,050 molL^{-1}$ de hidrogênio, e $0,050molL^{-1}$ de iodo, quais serão as concentrações dos reagentes e produtos no equilíbrio?

```python
from chempy import balance_stoichiometry
import chempy
```

```python
r1 = "H2"
r2 = "I2"
p = "HI"
```

```python
balance_stoichiometry({r1,r2},{p})
```

$$H_2+I_2 \leftrightarrows 2HI$$

$$k=\frac{[HI]^2}{[H_2][I_2]}$$

$$59.3=\frac{[-2x]^2}{[0.05-x][0.05-x]}$$

$$(0.05-x)^2 59.3 = 4x^2$$

$$x_1 \approx 0.04$$

$$x_2 \approx 0.07$$

```python
import sympy as sym
```

```python
x =  sym.symbols("x")
sym.init_printing("mathjax")
```

```python
e = 0.05-x
```

```python
e*e
```

```python
eqx = sym.decompose(((e*e)*59.3)-4*x**2)
eqx
```

```python
x1, x2 = sym.solve(eqx)
```

```python
x1
```

```python
x2
```

```python
# substituindo x1 nos reagentes 
e.subs(x,x1[x])
```

```python
# substituindo x2 nos reagentes
e.subs(x,x2[x])
```

Dessa forma, só pode ser usado o $x_1$

```python
produto = x1[x]*2
print(produto)
```

```python
H = e.subs(x,x1[x])
I = e.subs(x,x1[x])
```

```python
print(f"Resultado da concentração de Hidrogênio no equilibrio é: {H}")
```

```python
print(f"Resultado da concentração de Iodo no equilibrio é: {I}")
```

```python
print(f"Resultado da concentração de Ácido Iodídrico é : {produto}")
```
