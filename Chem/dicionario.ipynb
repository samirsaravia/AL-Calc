## Porcentagem da massa atômica 

```python
import re

# loop para calcular a massa atômica e a massa percentual atômica
x = ""
while x != '999':
    num_at = numero_atomico_massa
    x = input("\nColoque a fórmula molecular ou 999(sair): ")
    y = re.findall('[a-zA-Z][^A-Z]*', x)
    z = [re.split(r'(\d+)',s)[0:2] for s in (y)]  # lista para cada átomo como letra
    n = 0
    molar_mass = 0
    while len(z) > n:
        for formula in (z[n]):
            a = formula.split(',')  # converte numa lista
            b = " ".join(str(x) for x in a)
            try:
                c = float(b)
            except ValueError:
                atom = str(b)
                c = 1
                d =  num_at.get(atom)[1]
        molar_mass = molar_mass + (c * d)
        n = n + 1  #vai para o próximo átomo
        c = 1; d = 0
    print("Massa molar de ",x, " é: ", round(molar_mass,3))
    n = 0
    while len(z) > n:
        for formula in (z[n]):
            a = formula.split(',')
            b = " ".join(str(x) for x in a)
            try:
                c = float(b)
            except ValueError:
                atom = str(b)
                c = 1
                d = num_at.get(atom)[1]
        elemental_mass = (c * d)
        elemental_mass = (elemental_mass * 100/ molar_mass)
        print("Massa atômica % de ", atom, " é: ", round(elemental_mass, 3))
        n += 1
        c = 1 ; d = 0
```
