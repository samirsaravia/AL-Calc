# Equilibrio química em meio aquoso

Equilibrio de Solubilidade

```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as sym
from sympy.abc import x
```

```python
def f4(x,a,b,c,d,e):
    return a*x**4+b*x**3+c*x**2+d*x+e
```

# Ex 1
O produto de solubilidade do sulfato de prata $Ag_2SO_4$ em água a 25 graus, é $1.4x10^{-5}$
Estime a solubilidade molar do sal.



$$Ag_2SO_4 \leftrightarrows 2Ag^+ + SO_4^-$$
$K_{ps}= 1.4\times 10^{-5}$
$$K_{ps}= [2x]^2[x] = 4x^3$$
$$x=\sqrt[3] \frac{1.4\times 10^{-5}}{4} = 0.015$$
resultado é $0.015 mol L^{-1}$


## Gráfico

```python
# equação e integrada
K1 = 4*x**3-1.4e-5
Kint1 = sym.integrate(K1)
Kint1
```

```python
Kr1 = sym.solve(K1)
Kr1
```

```python
# vértice x=0.015
Kint1.subs(x,Kr1[0])
```

```python
# domínio e subdomínio
Kx1 = np.linspace(0.012,0.018,30)
Ky1 = f4(Kx1,1,0,0,-1.4e-5,0)
```

```python
# plotando os pontos cardinais encontrados
plt.plot(Kx1,Ky1)
plt.scatter(Kr1[0],Kint1.subs(x,Kr1[0]))
plt.show()
```

# Ex 2
O produto de solubilidade do fluoreto de chumbo(II) $PbF_2$, em água a 25˚C é $3,7.10^{-8}$. Estime a solubilidade molar do sal.

$$PbF_2 \leftrightarrows Pb^+ + 2F_2^-$$

$K_{ps}=3.7\times 10^{-8}molL^{-1}$


$$K_{ps}=[x][2x]^2= 4x^3$$

$$x = \sqrt[3]{\frac{3.7\times 10^{-8}}{4}}$$

$$x = 2.1 \times 10 ^{-3}$$

### Gráfico

```python
eq1 = 4*x**3-3.7e-8
req1 = sym.solve(eq1)
req1
```

```python
integrada = sym.integrate(eq1)
integrada
```

```python
M2 = integrada.subs(x,req1[0])
```

```python
Ax1 = np.linspace(-0.003,0.004,50)
Ay1 = f4(Ax1,1,0,0,-3.7e-8,0)
```

```python
plt.scatter(req1[0],M2)
plt.plot(Ax1,Ay1)
plt.show()
```

# Ex 3
A solubilidade molar do iodato de chumbo(II) $Pb(IO_3)_2$ em água a 25˚C é 40 $\micro mol L^{-1}$.
Qual é o valor de $K_{ps}$ do iodato de chumbo(II)?


$$Pb(IO_3)_2 \leftrightarrows Pb^+ +2 IO_3$$

$$K_{ps}= [x][2x]^2 = 4x^3 = 4(4\times 10^{-5})^3$$

$$K_{ps}=2.56\times 10 ^{-13}$$

Resultado é $K_{ps} = 2.56\times 10^{-13}molL^{-1}$
