```python
import matplotlib.pyplot as plt

# Dados de exemplo
resultados_teste = ['positivo', 'negativo', 'negativo', 'positivo', 'negativo', 'positivo']

# Gráfico de Barras
plt.bar(*zip(*dict((x, resultados_teste.count(x)) for x in set(resultados_teste)).items()))
plt.title('Resultados do Teste de COVID-19')
plt.xlabel('Resultado')
plt.ylabel('Frequência')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
veiculos = ['carro', 'moto', 'carro', 'caminhão', 'moto', 'carro', 'carro']

# Gráfico de Barras
plt.bar(*zip(*dict((x, veiculos.count(x)) for x in set(veiculos)).items()))
plt.title('Frequência dos Tipos de Veículos')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Frequência')
plt.show()
```
