# Tránsito

Por várias semanas, o serviço de trânsito vem pesquisando a velocidade de tráfego numa auto-estrada. Verifiou-se que, num dia normal de semana, à tarde entre 1 e 6 horas a velocidade do tráfego é de aproximadamente...
$$V(t)= 2t^3-21t^2+60t+40 \text{ km/h}$$ 

em que t é o número de horas transcorridas após o meio-dia. A que horas, dentro do intervalo de tempo mencionado, o tráfego se move mais rapidamente e que horas se move mais lentamente?

```python
import numpy as np
import sympy as sym
import math
import matplotlib.pyplot as plt
```

```python
x,y,t = sym.symbols("x y t")
```

```python
def funcao(x:list,a,b,c,d):
    return a*x**3+b*x**2+c*x+d
```

Gerando 50 número no Domínio entre 0,9 e 6,2 e sua respectiva Imagem, utilizando a equação $V(t)= 2t^3-21t^2+60t+40$

```python
# equação
velocidade = 2*t**3-21*t**2+60*t+40
```

```python
# Domínio e Imagem
x_lista = np.linspace(0.9,6.2,50)
y_lista = funcao(x_lista,2,-21,60,40)
```

Gerando o gráfico para os pontos

```python
plt.figure(num=0, dpi=100)
plt.grid(True, which="both")
plt.plot(x_lista, y_lista, label=f"$V(t)=2t^3-21t^2+60t+40$")
plt.legend()
plt.show()
```

 Calculando a variação Instantânea

```python
# derivada
aceleracao = sym.diff(velocidade)
aceleracao
```

Gerando Valores para o Domínio  e Imagem, utilizando a equação da aceleração em função do tempo:  $6t^2-42t+60$

```python
xt_lista = np.linspace(1.5,5.5,20)
yt_lista = funcao(xt_lista,0,6,-42,60)
```

Visualizando o gráfico para análise.

```python
plt.figure(num=0, dpi=120)
plt.grid(True, which="both")
plt.plot(x_lista, y_lista, label=f"$V(t)=2t^3-21t^2+60t+40$")
plt.plot(xt_lista, yt_lista, color="r", label=f"$a(t)=6t^2-42t+60$")
plt.legend()
plt.show()
```

Igualando a zero a equação da variação instantânea.

```python
def bhaskara(a,b,c):
    delta = (b*b) - (4*a*c)
    if delta < 0 :
        return "Não tem raizes reais"
    elif delta == 0:
        x = -b / (2*a)
        return x
    else:
        raiz = math.sqrt(delta)
        x1 = (-b + raiz)/(2*a)
        x2 = (-b - raiz)/(2*a)
        return [x1, x2]
```

```python
#6𝑡2−42𝑡+60
bhaskara(6,-42,60)
```

Temos duas raízes, $x_1= 2$ e $x_2=5$

Substituíndo esses valores, na equação Original, obtemos:

```python
velocidade.subs(t,2)   #(2,92)
```

```python
velocidade.subs(t,5)  #(5,65)
```

Encontrando as equações das retas nos pontos críticos.

```python
aceleracao.subs(t,5)
```

```python
# (5,65)
#y-65 = 0*(x - 5)
y = 65
```

```python
aceleracao.subs(t,2)
```

```python
# (2,92)
# y - 92 = 0(x-2)
y = 92
```

Atualizando o gráfico

```python
plt.figure(num=0, dpi=80)
plt.grid(True, which="both")
plt.plot(x_lista, y_lista, label=f"$V(t)=2t^3-21t^2+60t+40$")
#plt.plot(xt_lista, yt_lista, color="r", label=f"$a(t)=6t^2-42t+60$")
plt.plot((4,5,6),(65,65,65), color="g")
plt.plot((1,2,3),(92,92,92), color="g")
plt.scatter((2,5),(92,65),color="r", label="pontos max/min local")
plt.legend()
plt.show()
```

## Conclusão

Os resultado apontam que no intervalo da 1-6 horas o tráfego move-se mais rapidamente às 14hr, e mais lento às 17hr.
