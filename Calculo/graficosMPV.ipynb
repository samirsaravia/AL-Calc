```python
import matplotlib.pyplot as plt
from sympy import *
import numpy as np
```

```python
x_dominio = np.linspace(0,60,120)
```

```python
x = symbols("x")
```

```python
# equações todas de 3 grau
f_x = lambdify(x, -0.00176*x**3+0.15087*x**2-4.41282*x+95.9312)  # metal
f_x1 =lambdify(x,-0.00179*x**3+0.15809*x**2-4.78985*x+95.7684) # plastico
f_x2 = lambdify(x,-0.00161*x**3+0.14137*x**2-4.34331*x+96.4693)  #vidro
```

```python
# imagens
imagem_y = f_x(x_dominio)
imagem_plastico = f_x1(x_dominio)
imagem_vidro  = f_x2(x_dominio)
```

```python
# gráfico
plt.figure(figsize=(12,6))
plt.title("Resfriamento")
plt.grid(True, which="both")
plt.plot(x_dominio, imagem_y,label="metal")
plt.plot(x_dominio, imagem_plastico, label="plástico")
plt.plot(x_dominio, imagem_vidro, label="vidro")
plt.ylabel("Temperatura (C)")
plt.xlabel("Tempo (min)")
plt.ylim(20,101)
plt.legend()
plt.show()
```

```python
# grafico comparando
plt.figure(figsize=(12,6))
plt.title("Metal vs Plástico")
plt.grid(True, which="both")
plt.plot((x_dominio,x_dominio), (imagem_y, imagem_plastico))
plt.ylabel("Temperatura (C)")
plt.xlabel("Tempo (min)")
plt.ylim(20,101)
plt.show()
```

```python
# grafico comparando
plt.figure(figsize=(12,6))
plt.title("Metal vs Vidro")
plt.grid(True, which="both")
plt.plot((x_dominio,x_dominio), (imagem_y, imagem_vidro))
plt.ylabel("Temperatura (C)")
plt.xlabel("Tempo (min)")
plt.ylim(20,101)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Plástico vs Vidro")
plt.grid(True, which="both")
plt.plot((x_dominio,x_dominio), (imagem_y, imagem_vidro))
plt.ylabel("Temperatura (C)")
plt.xlabel("Tempo (min)")
plt.ylim(20,101)
plt.show()
```
