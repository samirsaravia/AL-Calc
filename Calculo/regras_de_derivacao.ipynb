```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
init_printing("mathjax")
```

```python
x,w,s,t, u = symbols("x w s t u")
```

## $f(x)= 3^x$

```python
a = 3**x
```

```python
diff(a,x)
```

```python
# gráfico 
dominio_a = np.linspace(0,5,20)
```

```python
fa = lambda x: 3**x
```

```python
f1a = lambdify(x, diff(a,x))
```

```python
imagem_a = fa(dominio_a)
```

```python
imagem_der_a = f1a(dominio_a)
```

```python
plt.grid(True, which="both")
plt.title("$f(x)=3^x$")
plt.plot(dominio_a, imagem_a, label="f(x)")
plt.plot(dominio_a, imagem_der_a, label="f'(x)")
plt.legend()
plt.show()
```

## $g(x)= x^4\times e^x$

```python
g = x**4 * exp(x)
g
```

```python
diff(g,x)
```

```python
g_eq = lambdify(x, x**4*exp(x).evalf())
g_1eq = lambdify(x, diff(g,x).evalf())
```

```python
dominio_g = np.linspace(-3,2, 40)
```

```python
imagem_g = g_eq(dominio_g)
imagem_g1 = g_1eq(dominio_g)
```

```python
plt.figure()
plt.grid(True, which= "both")
plt.title("$x^4 e^x$")
plt.plot(dominio_g, imagem_g, marker=".", label="g(x)")
plt.plot(dominio_g, imagem_g1,marker=".", alpha=0.5, label="g'(x)")
plt.legend()
plt.show()
```

## $h(x)= 2^x+5x$

```python
h = 2**x+5*x
h
```

```python
diff(h,x)
```

```python
dominio_h = np.linspace(-5,5,40)
```

```python
eq_h = lambdify(x, 2**x+5*x)
eq_h1 = lambdify(x, diff(h,x))
```

```python
imagem_h = eq_h(dominio_h)
imagem_h1 = eq_h1(dominio_h)
```

```python
plt.grid(True, which="both")
plt.plot(dominio_h, imagem_h, label="h(x)")
plt.plot(dominio_h, imagem_h1, label="h'(x)")
plt.legend()
plt.show()
```
