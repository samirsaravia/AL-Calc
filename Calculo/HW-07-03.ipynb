```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt 
```

```python
#x = smp.symbols("x")
def main():
    global x
    x = smp.symbols("x")
    x_ls = np.linspace(-4,2,100)
    #x_ls = np.arange(-4,2,.1)
    y_ls = g(x_ls, 1,2,7) 
    grafico(x_ls,y_ls,"x^2+2x+7")
    print(bhaskara(1,2,7))

    
def grafico(xlista:list, ylista:list, formula:str) -> None:
    plt.figure(num=0, dpi=120)
    plt.plot(xlista, ylista, label=f"${formula}$")
    plt.legend()
    plt.ylim(-2,max(ylista)+1)
    plt.grid(True, which="both")
    #plt.savefig("mat.svg",format="svg")
    plt.show()
    
    
def g(x,a,b,c):
    return a*x**2 + b*x + c


def bhaskara(a,b,c):
    delta = (b*b) - (4*a*c)
    if delta < 0 :
        return "Não tem raizes reais"
    elif delta == 0:
        x = -b / (2*a)
        return x
    else:
        raiz = sqrt(delta)
        x1 = (-b + raiz)/(2*a)
        x2 = (-b - raiz)/(2*a)
        return [x1, x2]
```

```python
if __name__ == "__main__":
    main()
```

## Vertices

```python
def vertices(a,b,c):
    x_vertice = -b/2*a
    y_vertice = -(b**2 - (4*a*c))/4*a
    return [x_vertice,y_vertice]

vertices(1,2,7)
```
