# Caneca de Vidro

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model  import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sympy import *
```

```python
#x =  symbols("x")
init_printing("mathjax")
```

```python
for i in range(1,6):
    #dominio e imagem
    y = np.array([99.8,72.7,63.9,59.5,56.5,52.6,48.8,46.5,44.5,42.3]).reshape(-1,1)
    x = np.array([0,5,10,15,20,25,30,35,40,45]).reshape(-1,1)

    #treinamento ML
    poly_features = PolynomialFeatures(degree=i, include_bias=False)
    X_poly = poly_features.fit_transform(x)
    model =  LinearRegression()
    model.fit(X_poly,y)
    X_vals = np.linspace(-1,50,20).reshape(-1,1)
    X_vals_poly = poly_features.transform(X_vals)
    y_vals = model.predict(X_vals_poly)


    #equação
    vind = model.intercept_
    coeficiente = model.coef_

    coef_lista = []
    for k in coeficiente:
        for j in k:
            coef_lista.append(f"{j:.5f}")

    
    #grafico
    plt.grid(True, which="both")
    plt.scatter(x,y)
    plt.plot(X_vals,y_vals, color="r")
    plt.title(f"{i}° Grau")
    #plt.savefig(f"{grau}° grau")
    plt.xlabel("Tempo (min)")
    plt.ylabel("Temperatura (C°)")
    plt.show()

    #equação formatada em listas
    print(f"Coeficientes para equação, (x⁰,...,x^{i})")
    print(f"{vind[0]:.4f}",coef_lista)
```
