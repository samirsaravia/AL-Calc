## Cálculos usando símbolos

As vezes a formula não é tão simples de resolver.
$$f(x)= ...$$

**Exemplo:**

$$f(x)=e^{-a \sin (x^2)}\cdot sin(b^x) \cdot \ln(c \sin x^2/x)$$

```python
x,a,b,c = sym.symbols('x a b c', real=True)
f = sym.exp(-a*sym.sin(x**2))*sym.sin(b**x)*sym.log(c*sym.sin(x)**2/x)
```

```python
f
```

Para calcular as derivadas usa-se `sym.diff(f,x)` onde $f$ é a função para derivar e $x$ é a variável a qual está derivando.

```python
dfdx = sym.diff(f,x)
dfdx
```

É possível determinar a $n$ derivada $d^n \frac{f}{dx^n}$ agregando o argumento opcional 
`sym.diff(f,x,n)`

```python
d4fdx4 = sym.diff(f,x,4)
# 
```

Pode-se computar valores numéricos adicionando estes

```python
d4fdx4.subs([(x,4),(a,1),(b,2),(c,3)]).evalf()  #evalf serve para calcular (mostrar um valor)
```

Pode-se converter à uma função para plotar

```python
d4fdx4_f = sym.lambdify((x,a,b,c),d4fdx4)  #isso aqui é muito poderoso
type(d4fdx4_f)
```

```python
dominio = np.linspace(1,2,50)
imagem = d4fdx4_f(dominio, a=1,b=2,c=3)
```

```python
# Plot
plt.plot(dominio, imagem, 'o--')
plt.ylabel('$d^4f/dx^4$', fontsize=24)
plt.xlabel('$x$', fontsize=24)
plt.show()
```
