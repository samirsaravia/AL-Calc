## $g(x)= x \cos (x)$

```python
g1 = x* cos(x)
g1
```

```python
# Resposta
diff(g1,x)
```

```python
x1 = np.linspace(-1,5,30)
```

```python
d1 = lambdify(x, diff(g1,x))
y1 = d1(x1)
d11 = lambdify(x, g1)
y11 = d11(x1)
```

```python
def grafico(imagem, dominio1, dominio2,marker=False):
    if marker:
        plt.figure()
        plt.grid(True, which="both")
        plt.plot(imagem, dominio1, label="g(x)",marker=".")
        plt.plot(imagem, dominio2, label="g'(x)", marker=".")
        plt.legend()
        plt.show()
    else:
        plt.figure()
        plt.grid(True, which="both")
        plt.plot(imagem, dominio1, label="g(x)")
        plt.plot(imagem, dominio2, label="g'(x)")
        plt.legend()
        plt.show()
    
```

```python
grafico(x1, y1, y11)
```

## Estudo do café
No estudo do comportamento da temperatura de
100 mL de café em uma caneca de alumínio
esmaltado à temperatura ambiente de 28 0C,
deduziu-se o seguinte modelo matemático
$$T(t)=\frac{23.59}{1-0.61e^{-0.02t}}$$

Calcule a derivada da temperatura do café em função do tempo. O que essa derivada representa?

```python
T1 = 23.59/(1-0.61*exp(-0.02*t))
T1
```

```python
T11 = diff(T1, t)
T11
```

```python
T1x = np.linspace(9, 21,50)
```

```python
T1f = lambdify(t, T1)
T11f = lambdify(t,T11)
T1y = T1f(T1x)
T11y = T11f(T1x)
```

```python
grafico(T1x, T1y, T11y)
```

Calcule a derivada da temperatura do café para o tempo de 10 minutos. O valor obtido é positivo
ou negativo? O que isso representa?

```python
T11.subs(t,10)
```

Calcule a derivada da temperatura do café para o tempo de 20 minutos. O valor obtido é maior
ou menor do que o obtido no item b? O que isso representa?

```python
T11.subs(t,20)
```
