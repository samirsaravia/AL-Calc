# Trajeto de Rolandia a Londrina

Um grupo de alunos, com interesse de representar por meio de um modelo matemático as velocidades no trajeto de Rolândia (casa de um dos integrantes) a Londrina (até a UTFPR), realizou a viagem em uma quarta-feira, entre as 18h e 19h.
Durante o trajeto que se iniciou da casa de um dos alunos até a UTFPR, o grupo observou o painel do carro e anotou a velocidade em períodos de 3 em 3 minutos, ou seja, 0,05 em 0,05 de hora. Para isso, organizou os dados coletados em uma tabela.Determine a distância aproximada da casa do aluno até à UTFPR>
 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model  import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import sympy as sym
```

```python
x = np.array([0,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45,0.50,0.55,0.60]).reshape(-1,1)
y = np.array([0,50,80,100,82,80,80,35,40,25,40,40,45]).reshape(-1,1)
```

```python
grau = 4
poly_features = PolynomialFeatures(degree=grau, include_bias=False)
```

```python
X_poly = poly_features.fit_transform(x)
model =  LinearRegression()
model.fit(X_poly,y)
```

```python
X_vals = np.linspace(0,0.6,100).reshape(-1,1)
X_vals_poly = poly_features.transform(X_vals)
y_vals = model.predict(X_vals_poly)
```

Gráfico

```python
plt.grid(True, which="both")
plt.scatter(x,y)
plt.plot(X_vals,y_vals, color="r")
plt.title(f"{grau}° Grau")
#plt.savefig(f"{grau}° grau")
plt.show()
```
