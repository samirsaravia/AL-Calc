# Uso simples de matrizes em encriptar imagens

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Testando para usar matrizes

```python
v1 = np.array([4,4])
v2 = np.array([-1,-2])
```

```python
#plot points
plt.scatter([0,v1[0],v1[0]+v2[0]],[0,v1[1],v1[1]+v2[1]], color="red")

#plot line
plt.plot([0,v1[0],v1[0]+v2[0]],[0,v1[1], v1[1]+v2[1]])

plt.grid(True, which='both')
plt.xticks(np.arange(-2,7,1))
plt.yticks(np.arange(-2,7,1))
plt.axvline(x=0, color="black")
plt.axhline(y=0, color="black")
plt.title("Matrix addition")
plt.show()
```

## Encriptando uma imagem

A função `imread` do matplotlib consegue ler dados de uma imagem numa matriz em Numpy. Os pixeis, são representados como 8bit, inteiros sem sinal (positivo ou  negativo) dentro de uma matriz.

Imagem selecionada no pexels (small 640x427) [Imagem](https://www.pexels.com/photo/macro-photography-of-dragonfly-on-plant-2918152/)

```python
# ler a foto e mostrar a ordem da matriz
foto = plt.imread("pexels-kelly-1179532-2918152.jpg")
foto.shape
```

```python
plt.imshow(foto)
```

```python
numero_primo = [] # usado para valores numéricos primos
for numero in  range(2,255):
    if all(numero%i != 0 for i in range(2,numero)):
        numero_primo.append(numero)

len(foto)
```

```python
numero_primo = numero_primo * len(foto)
np.random.shuffle(numero_primo)
numeros_baralhados = np.random.randint(0,255, size=len(foto))
```

```python
print(numero_primo)
```

```python
numeros_baralhados
```

```python
foto_encriptada = np.array([(x*(x^y)) % z 
                            for x,y,z in zip(foto, numeros_baralhados, numero_primo)])
```

```python
plt.imshow(foto_encriptada)
plt.show()
```
