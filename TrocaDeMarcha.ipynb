```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([7.5,9,10.4,9.1,7.8]).reshape(-1,1)
```

```python
def f(x,a,b,c):
    return a*x**2 + b*x + c
```

```python
#xlist = np.linspace(0,6,num=100)  #num gera uma quatidade de dados entre os valores estabelecidos
xlist = np.arange(0,6.1,.1)
print(xlist)
print(len(xlist))
```

```python
ylist = f(xlist, -0.593, 3.627, 4.400)
ylist
```

```python
# mostra os pontos (x,y)
plt.scatter(x,y, color="r")

#plt.figure(num=0,dpi=120)
plt.title(r"RPM/Km")
plt.plot(xlist, ylist, label="Legend")
plt.xlim(0,7)
plt.grid(True, which="both", label="THIS")

#plota etiquetas
plt.xlabel("RPM")
plt.ylabel("Km")

#plota legendas
plt.legend()

#plota tudo
plt.show()
```
